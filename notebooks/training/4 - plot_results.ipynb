```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def plot_measures(x, a, b, label_a, label_b, label_x, label_y, title):
    plt.figure(figsize=(3, 4))
    plt.plot(x, a, label=label_a)
    plt.plot(x, b, label=label_b)
    plt.xlabel(label_x)
    plt.ylabel(label_y)
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# PER YOLO 5
# Leggi il file results.csv
df = pd.read_csv('../../detection/weights/yolo v5/gram/exp/results.csv')
df.columns = df.columns.str.strip()

print("Epoche:", len(df['epoch']))

# Grafici delle Loss
plot_measures(df['epoch'], df['train/box_loss'], df['val/box_loss'], 'Train Box Loss', 'Validation Box Loss', 'Epoch', 'Loss', 'Loss per Epoch')
plot_measures(df['epoch'], df['train/cls_loss'], df['val/cls_loss'], 'Train Class Loss', 'Validation Class Loss', 'Epoch', 'Loss', 'Loss per Epoch')
plot_measures(df['epoch'], df['train/obj_loss'], df['val/obj_loss'], 'Train Object Loss', 'Validation Object Loss', 'Epoch', 'Loss', 'Loss per Epoch')

# Grafico di Precision e Recall
plot_measures(df['epoch'], df['metrics/precision'], df['metrics/recall'], 'Precision', 'Recall', 'Epoch', 'Value', 'Precision and Recall per Epoch')

# Grafico di mAP
plot_measures(df['epoch'], df['metrics/mAP_0.5'], df['metrics/mAP_0.5:0.95'], 'mAP@0.5', 'mAP@0.5:0.95', 'Epoch', 'mAP', 'mAP per Epoch')
```
