Disegna l'area di tracking di un frame del video.

Formato 1920 x 1080

```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

main_path = "../../"
```

```python
video_file = "IMG_0506.MOV"
test_video = os.path.join(main_path, "datasets/thai/Videos/", video_file)
```

```python
cap = cv2.VideoCapture(test_video)
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    break
cap.release()
frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

overlay = frame_rgb.copy()

cv2.line(frame_rgb, (0, 200), (1920, 200), (255, 0, 0), 2)
cv2.line(frame_rgb, (0, 800), (1920, 800), (255, 0, 0), 2)

x, y, w, h = 0, 200, 1920, 600
cv2.rectangle(overlay, (x, y), (x+w, y+h), (0, 255, 0), -1)
alpha = 0.4

white_rect = np.ones(frame_rgb.shape, dtype=np.uint8) * 255
trasp = (0, 255, 0, 125)
image_new = cv2.addWeighted(overlay, alpha, frame_rgb, 1 - alpha, 0)

plt.figure(figsize=(16, 9))
plt.imshow(image_new)
plt.axis('off')
plt.show()
```
