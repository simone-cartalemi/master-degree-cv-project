Ottieni la trasformazione omografica di un frame del video.

Formato 1920 x 1080

```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

main_path = "../../"
```

```python
video_file = "IMG_0506.MOV"
test_video = os.path.join(main_path, "datasets/thai/Videos/", video_file)
matrix_path = os.path.join(main_path, "homography/matrixes/thai_hom_matrix_last.npy")

M = np.load(matrix_path)
```

```python
cap = cv2.VideoCapture(test_video)
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    break
cap.release()
frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

height, width, channels = frame_rgb.shape

warped_image = cv2.warpPerspective(frame_rgb, M, (width, height))

# Frame originale
plt.figure(figsize=(16, 9))
plt.imshow(frame_rgb)
plt.axis('off')
plt.show()

# Frame rimappato
plt.figure(figsize=(16, 9))
plt.imshow(warped_image)
plt.axis('off')
plt.show()
```
